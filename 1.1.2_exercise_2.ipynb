```python
import altair
import pandas
import numpy
from scipy import stats
import scipy
```

Using three different distribution classes in scipy.stats, run the following experiment:

1. Compute 10,000 random values. Compute and save the mean.
2. Repeat step 1, 1,000 times
3. Plot the histogram of the mean values and observe the shape.

Use the distributions norm, cosine and uniform.


```python
norm_dist = stats.norm()
cosine_dist = stats.gamma(0.5)
uniform_dist = stats.uniform(0, 1)



means = []
for y in range(1000):
    values = norm_dist.rvs(10000)
    mean = scipy.mean(values)
    means.append(mean)
df_values = pandas.DataFrame({"value": means})
plot = altair.Chart(df_values).mark_bar().encode(
    altair.X("value", bin=altair.Bin(maxbins=100)),
    y='count()'
)
plot

    
```
